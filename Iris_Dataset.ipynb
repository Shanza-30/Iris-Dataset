```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("Iris.csv")
```

```python
# Summary statistics
df.describe()
```

```python
# Pairplot scatter
sns.pairplot(df, hue="Species", markers=["o", "s", "D"])
plt.suptitle("Scatter Plot (Pairplot) of Features", y=1.02)
plt.show()
```

```python
# Histograms for numeric columns
df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(12, 12)) # Adjusted figsize to accommodate more plots
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(3, 2, i) # Changed subplot grid from 2x2 to 3x2
    sns.boxplot(x="Species", y=column, data=df)
    plt.title(f"Box Plot of {column}")

plt.tight_layout()
plt.show()
```
